#importing datasets

#Data pre-processing


```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le6=LabelEncoder()
le8=LabelEncoder()
le10=LabelEncoder()
X[:,1]=le1.fit_transform(X[:,1])
X[:,2]=le2.fit_transform(X[:,2])
X[:,6]=le6.fit_transform(X[:,6])
X[:,8]=le8.fit_transform(X[:,8])
X[:,10]=le10.fit_transform(X[:,10])
```
